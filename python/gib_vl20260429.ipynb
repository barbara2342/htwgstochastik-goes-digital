```python
# Demo file for working with simple random variables
# live entstanden in der Vorlesung GIB Mathematik2, 29.04.2026
#
# copyright: Barbara Staehle, HTWG Konstanz
# bstaehle@htwg-konstanz.de

# import libraries we use for plotting and calculating
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
```

```python
# Beispiel ZV X, Gewinn beim Spiel mit einem einarmigen Banditen (siehe Skript)
x = [-1, 4, 9, 14,19]
p = [0.977, 0.008, 0.008, 0.006, 0.001]

print('Y-Achse linear skaliert: verdeutlicht hohe Wahrscheinlichkeit des Verlusts, macht es schwer andere Wahrscheinlichkeiten zu unterscheiden')
print(' ') 
plt.bar(x,p)
plt.grid(True)
plt.xlabel('Gewinn x')
plt.ylabel('Wahrscheinlichkeit des Gewinns, P(X = x)')
plt.yscale('linear')
plt.xticks(x)
plt.title('Wahrscheinlichkeitsverteilung der ZV X, Gewinn')
plt.show()

```

```python
print('Y-Achse logarithmisch skaliert: verdeutlicht Unterschiede der Gewinn-Wahrscheinlichkeiten, verniedlicht hohe Verlustwahrscheinlichkeit')
print(' ') 

plt.bar(x,p)
plt.grid(True)
plt.xlabel('Gewinn x')
plt.ylabel('Wahrscheinlichkeit des Gewinns, P(X = x)')
plt.yscale('log')
plt.xticks(x)
plt.title('Wahrscheinlichkeitsverteilung der ZV X, Gewinn')
plt.show()

```
