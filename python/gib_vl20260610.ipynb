```python
# Beispiel aus dem Skript: Die Zeit die zwischen zwei Anforderungen in einem Mobilfunknetz wird durch eine exponantialverteilte Zufallsvariable mit λ = 0.5 sec−1 beschrieben. 

lam = 0.5 #(NIEMALS einen Parameter lamba nennen, eine Lambda-Funktion eine anonyme Funktion: https://python-basics-tutorial.readthedocs.io/de/latest/functions/lambda.html

# expon hat zwei Parameter: 
# loc (gibt die Verschiebung vom Ursprung an, für unsere Anwendungen ist das immer 0)
# scale gibt die Rate an. 
# 
# Doku: A common parameterization for expon is in terms of the rate parameter lambda, such that pdf = lambda * exp(-lambda * x). This parameterization corresponds to using scale = 1 / lambda
# => in unserem Beispiel muss scale = 1/λ = 1/0.5 = 2 sein!

A = expon(loc = 0, scale = 1/lam)
```

```python
# Wie groß ist die Wahrscheinlichkeit, dass zwischen zwei Ankünften höchsten 1 Sekunde ist : P(A <= 1)

# schlampig und schnell - hier muss man aber darauf achten, die Parameter 0 UND 2 (loc und scale), sowie die Stelle, an der die cdf berechnet werden soll anzugeben
expon.cdf(1,0,2)
```

```python
# # über das Objekt der ZV A, wenn man es eh schon angelegt hat
A.cdf(1)
```

```python
# händisch über die Formel P(A <= 1) = 1 - e^(-1*lam)
1 - np.exp(-lam)
```

```python
# ... zwischen 1 und 2 Sekunden liegt: P(1 < A < 2)

# schlampig und schnell
expon.cdf(2,0,2) - expon.cdf(1,0,2)
```

```python
# über das Objekt der ZV A
A.cdf(2)-A.cdf(1)
```

```python
# händisch über die Formel: P(1 < A < 2) = e^(-lam)-e^(-lam*2)
np.exp(-0.5)-np.exp(-1)
```

```python
# ... mehr als 2 Sekunden: P(A > 2) = 1 - P(A <= 2)

# über das Objekt der ZV A
1-A.cdf(2)
```

```python
# über das Objekt der ZV A mit survial function
A.sf(2)
```

```python
# schlampig und schnell
1 - expon.cdf(2,0,2)
```

```python
# händisch über die Formel:  P(A > 2) = 1 - P(A <= 2) = e^(-2*lam)
np.exp(-1)
```

```python
# Nach welcher Zeit kommt mit 60%-iger Wahrscheinlichkeit spätestens wieder eine Anforderung im Netz von P3 an?
A.ppf(0.6)
```

```python
# Geben Sie den Median von A an

# über Median
A.median()
```

```python
# als 50-Prozent-Quantil
A.ppf(0.5)
```
