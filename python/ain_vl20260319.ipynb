```python
# Demo file for working with univariate data
# live entstanden in der Vorlesung AIN Stochastik, 19.03.2026
#
# copyright: Barbara Staehle, HTWG Konstanz
# bstaehle@htwg-konstanz.de

# import libraries we use for plotting and calculating
import matplotlib.pyplot as plt
import numpy as np 
```

```python
# body length in cm of students attending the lecture
x = [170, 161, 173, 176, 172, 189, 167, 173]
print(x)
```

```python
# plot koerpergroessen
plt.figure()
# this results in Teilnehmernummern 0-N-1
#plt.plot(x, marker='*', markersize=5, linestyle='none',linewidth=5,color='green')
## range(1,len(x)+1) cares for having 1-N on the x-axis
plt.plot(range(1,len(x)+1),x, marker='*', markersize=5, linestyle='none',linewidth=5,color='green')
# label your axis!!
plt.xlabel('Teilnehmernummer')
plt.ylabel('Körpergröße [cm]')
# Begrenzung der y-Achse ist Geschmackssache
plt.ylim(150,200)
plt.grid(True)
plt.show()
```

```python
# count how often each value occurs
values, counts = np.unique(x,return_counts=True)
print(f'Werte in der Liste: {values}')
print(f'Absolute Häufigkeiten: {counts}')
```

```python
# show as bar plot
plt.bar(values,counts)
plt.grid(True)
plt.xlabel('Körpergröße [cm]')
plt.ylabel('absolute Häufigkeit')
plt.yticks(range(0,max(counts)+1))
plt.show()
```

```python
# show as stem plot
plt.stem(values,counts)
plt.grid(True)
plt.xlabel('Körpergröße [cm]')
plt.ylabel('absolute Häufigkeit')
plt.yticks(range(0,max(counts)+1))
plt.show()
```

```python
plt.title('Achtung, so bitte auf keinen Fall darstellen!')
plt.plot(values,counts,color='r')
plt.grid(True)
plt.xlabel('Körpergröße [cm]')
plt.ylabel('absolute Häufigkeit')
plt.yticks(range(0,max(counts)+1))
plt.title('Achtung, so bitte auf keinen Fall darstellen!')
plt.show()
```

```python
# show graphically via dots only (do not connect)
plt.plot(values,counts,marker='*',linestyle='none')
plt.grid(True)
plt.xlabel('Körpergröße [cm]')
plt.ylabel('absolute Häufigkeit')
plt.yticks(range(0,max(counts)+1))
plt.title('Diese Darstellung kann man machen, ich finde sie aber nicht so toll')
plt.show()

```

```python
# empirische kumlulierte Häufigkeiten mit ecdf (schnell aber unhübsch)
plt.ecdf(x)
plt.xlabel('Körpergröße [cm]')
plt.ylabel('Kumulierte Häufigkeit')
plt.grid(True)
plt.title('unhübsch, weil am Anfang die 0 fehlt, am Ende die 1')
plt.show()
```

```python
# mit Funktionen cumsum und step (aufwändiger aber hübscher)
x = np.array(x)
x2 = np.append(x, [min(x)-5,max(x)+5])
y = np.array(range(1,len(x)+1))/len(x)
y = np.append(y, [0, 1])
plt.step(sorted(x2),sorted(y),where='post')
plt.xlim(min(x)-2,max(x)+2)
plt.grid(True)
plt.xlabel('Körpergröße [cm]')
plt.ylabel('Kumulierte relative Häufigkeit')
plt.title('Hübscher und mathematisch korrekter')
plt.show()
```
