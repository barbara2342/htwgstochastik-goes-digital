```python
# Demo file for working with simple random variables
# live entstanden in der Vorlesung AIN Stochastik, 07.05.2026
#
# copyright: Barbara Staehle, HTWG Konstanz
# bstaehle@htwg-konstanz.de

# import libraries we use for plotting and calculating
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

```

```python
# Beispiel ZV M und S, Klausurnotenwürfel der Professor:innen Müller und Schmidt (siehe VL)

# die Werte (Noten von 1-5) und Wahrscheinlichkeiten können nicht verändert werden, verwende daher hier Tupel (runde Klammern, schneller) statt Listen (eckige Klammern) 
x = np.arange(1,6)
ps = (2/6, 3/6, 1/6, 0, 0)
pm = (0, 1/6, 2/6, 2/6, 1/6)

# Lege M und S als diskrete Zufallsvariablen an
# https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.rv_discrete.html
M = stats.rv_discrete(name='custm', values=(x, pm))
S = stats.rv_discrete(name='custm', values=(x, ps))
```

```python
# Plotte Wahrscheinlichkeitsverteilung 

plt.bar(x-0.1, M.pmf(x),width=0.3,label='M')
plt.bar(x+0.1, S.pmf(x),width=0.3,label='S')
plt.xlabel('Noten (x)')
plt.ylabel('Wahrscheinlichkeit  (P(M=x), P(S=x))')
plt.title('Wahrscheinlichkeitsverteilung der ZV M und S')
plt.legend()
plt.grid(True)

plt.show()
```

```python
# Plotte kumulative Wahrscheinlichskeitsverteilungsfunktion
# verwende x2 = 0, 1, ... 6 statt 1, 2, ... 5 für eine hübscher Darstellung
x2 = range(0,7)
# verwende where='post', sonst wird der plot falsch 
plt.step(x2, M.cdf(x2),where='post',label='M')
plt.step(x2, S.cdf(x2),where='post',label='S')
plt.legend()
plt.grid(True)
plt.xlabel('Noten (x)')
plt.ylabel('Kumlulierte Wahrscheinlichkeit  (P($M\leq x$), P($S\leq x$))')
plt.title('Kumulierte Wahrscheinlichkeitsverteilungsfunktion der ZV M und S')

plt.show()
```

```python
# Erroneous plot!!
# use originial x-Werte (1, 2, ... 5)

# omit "where='post'" -> default "where='pre'" is used 
plt.step(x, M.cdf(x),label='M')
plt.step(x, S.cdf(x),label='S')
plt.legend()
plt.grid(True)
plt.xlabel('Noten (x)')
plt.ylabel('Kumkulierte Wahrscheinlichkeit  (P($M\leq x$), P($S\leq x$))')
plt.title('Kumulierte Wahrscheinlichkeitsverteilungsfunktion der ZV M und S')
plt.text (2.5,0.3,'Achtung, Darstellung fehlerhaft!!',color='red',fontweight='bold',backgroundcolor='white',bbox=dict(boxstyle='round', facecolor='white',edgecolor='red'))

plt.show()
```

## Mittelwert, Varianz, Standardabweichung

```python
# Mittels Funktionalität der ZV-Bibliothek
print(f'E(M): {M.mean()}')
print(f'E(S): {S.mean():.3f}')
```

```python
# von Hand
print(f'E(M): {np.dot(pm,x.T)}')
print(f'E(S): {np.dot(ps,x.T):.3f}')
```

```python
# Mittels Funktionalität der ZV-Bibliothek
print(f'Var(M): {M.var():.3f}')
print(f'Var(S): {S.var():.3f}')
```

```python
# Mittels Funktionalität der ZV-Bibliothek
print(f'Std(M): {M.std():.3f}')
print(f'Std(S): {S.std():.3f}')
```

```python
# von Hand
import math 

print(f'Std(M) als Wurzel: {math.sqrt(M.var()):.3f}')
print(f'Std(S) als Wurzel: {math.sqrt(S.var()):.3f}')
```
