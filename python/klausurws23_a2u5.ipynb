```python
# 5.2f
# symmetrischen Intervall in dem 95% aller Fälle liegen, heißt der Rest muss sich symmetrisch auf unteren und oberen Randbereich aufteilen
rest = 1-0.95
xu = rest/2
xo = 1-rest/2
```
